# Q10  Using variables

Assign two variables `a = 15` and `b = 4`. Compute and print **all** arithmetic operations between them: `+`, `-`, `*`, `/`, `//`, `%`, `**`.

```python
a = 15
b = 4

print(a + b)   # Addition
print(a - b)   # Subtraction
print(a * b)   # Multiplication
print(a / b)   # Division
print(a // b)  # Floor division
print(a % b)   # Modulus (remainder)
print(a ** b)  # Power
```

# Q11 - Compound assignment

Start with `n = 10`. Use compound assignment operators to:
- add 5 (`n += 5`), print
- multiply by 2 (`n *= 2`), print
- subtract 4 (`n -= 4`), print
(Place each statement so students can see the updated `n` after each operation.)

```python
n = 10

n += 5    # Add 5
print(n)  # Output: 15

n *= 2    # Multiply by 2
print(n)  # Output: 30

n -= 4    # Subtract 4
print(n)  # Output: 26
```

# Q12  Swapping values (without a third variable)

Given `x = 3` and `y = 7`, swap their values using Python tuple assignment and print `x` and `y` after swapping.

```python
x = 3
y = 7

x, y = y, x  # Swap values
print(x)     # Output: 7
print(y)     # Output: 3

```

# Q15 - Input: Circle calculations

Ask the user to input the radius of a circle (float). Compute and print:
- Circumference = `2 * 3.14159 * radius`

```python
# Ask the user to input the radius
radius = float(input("Enter the radius of the circle: "))

# Compute circumference and area
circumference = 2 * 3.14159 * radius
area = 3.14159 * radius ** 2

# Print the results
print("Circumference:", circumference)
print("Area:", area)
```
