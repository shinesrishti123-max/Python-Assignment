## Question 2 (Easy)

**Title:** Sum of First N Numbers

### Problem Statement:
Take an integer N as input and find the sum of numbers from 1 to N.

### Approach:
Initialize sum = 0, loop from 1 to N, add each value to sum.


```python
N = int(input("Enter a number N: "))
total = 0

for i in range(1, N + 1):
    total += i

print("Sum of first", N, "numbers is:", total)

```

## Question 9 (Easy)

**Title:** Count Even and Odd Numbers

### Problem Statement:
Take N as input and count how many even and odd numbers are there between 1 and N.

### Approach:
Use two counters and check with modulo operator.


```python
N = int(input("Enter a number N: "))
even_count = 0
odd_count = 0

for i in range(1, N + 1):
    if i % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

print(f"Even numbers count: {even_count}")
print(f"Odd numbers count: {odd_count}")
```

## Question 13 (Medium)

**Title:** Skip Multiples of 4

### Problem Statement:
Take N as input and print numbers from 1 to N but skip multiples of 4.

### Approach:
Use `continue` when the number is divisible by 4.


```python
N = int(input("Enter a number N: "))

for i in range(1, N + 1):
    if i % 4 == 0:
        continue  # Skip multiples of 4
    print(i, end=" ")
```

## Question 14 (Medium)

**Title:** Find the Largest Number

### Problem Statement:
Take N as input and find the largest number between 1 and N using a loop.

### Approach:
Assume the first value as largest and update using `if`.


```python
N = int(input("How many numbers do you want to enter? "))
largest = None

for i in range(N):
    num = int(input(f"Enter number {i+1}: "))
    if largest is None or num > largest:
        largest = num

print(f"The largest number is: {largest}")
```

## Question 15 (Medium)

**Title:** Power of a Number

### Problem Statement:
Take two numbers `a` and `b` as input and calculate `a` raised to the power `b` using a for loop.

### Approach:
Initialize result = 1 and multiply it `b` times.


```python
a = int(input("Enter base number a: "))
b = int(input("Enter exponent b: "))
result = 1

for _ in range(b):
    result *= a

print(f"{a} raised to the power {b} is: {result}")
```

## Question 16 (Hard)

**Title:** Check Prime Number

### Problem Statement:
Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
n = int(input("Enter a number: "))

if n <= 1:
    print("Not a prime number")
else:
    for i in range(2, n):
        if n % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime number")
```

## Question 17 (Hard)

**Title:** Prime Numbers Between 1 and N

### Problem Statement:
Take N as input and print all prime numbers between 1 and N.

### Approach:
Use nested loops and a flag variable for prime checking.


```python
N = int(input("Enter N: "))

for num in range(2, N + 1):   
    flag = 1                 

    for i in range(2, num):  
        if num % i == 0:
            flag = 0        
            break

    if flag == 1:
        print(num, end=" ")
```

Take N as input and print a right-angled number pattern using nested loops.

Example:

## Question 20 (Hard)

**Title:** Perfect Number Check

### Problem Statement:
Take a number as input and check whether it is a perfect number or not.

### Approach:
Find sum of all factors using a loop and compare with the number.


```python
n = int(input())

s = 0
for i in range(1, n):
    if n % i == 0:  
        s += i      

if s == n:
    print("Perfect Number")
else:
    print("Not a Perfect Number")
```
