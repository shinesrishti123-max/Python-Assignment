## Problem 2: Leap Year

**Question:** Check if a given year is a leap year.

**Approach:** Apply the rules: divisible by 4, but centuries must be divisible by 400.

Rules: Year divisible by 4 is leap, but if divisible by 100 it must also be divisible by 400.

```python
# Enter your answer here
year = int(input("Enter a year: "))

if (year % 400 == 0):
    print("Leap year")
elif (year % 100 == 0):
    print("Not a leap year")
elif (year % 4 == 0):
    print("Leap year")
else:
    print("Not a leap year")

```

## Problem 3: Grade Bands

**Question:** Print grade category based on marks: Excellent, Good, Average, Fail.

**Approach:** Build multiple conditions using elif to check ranges.

Take marks (0-100) and print:
- 90–100 → Excellent
- 75–89 → Good
- 50–74 → Average
- else → Fail

```python
marks = int(input("Enter marks (0-100): "))

if 90 <= marks <= 100:
    print("Excellent")
elif 75 <= marks <= 89:
    print("Good")
elif 50 <= marks <= 74:
    print("Average")
else:
    print("Fail")
```

## Problem 7: Delivery Charge System

**Question:** Calculate final delivery charge based on prime membership, amount, and distance.

**Approach:** Start with prime-check, then inside it apply nested if blocks.

Inputs:
- order_amount (int)
- is_prime_member (yes/no)
- distance_km (float)

Rules:
- If prime member → Free delivery
- Else:
  - If amount ≥ 500:
    - If distance ≤ 5 km → Free
    - Else → ₹20 extra
  - Else:
    - If distance ≤ 5 km → ₹30
    - Else → ₹50

```python
order_amount = int(input("Enter order amount: "))
is_prime = input("Prime member? (yes/no): ").lower()
distance = float(input("Enter distance in km: "))

if is_prime == "yes":
    print("Delivery Charge: Free")
else:
    if order_amount >= 500:
        if distance <= 5:
            print("Delivery Charge: Free")
        else:
            print("Delivery Charge: ₹20")
    else:
        if distance <= 5:
            print("Delivery Charge: ₹30")
        else:
            print("Delivery Charge: ₹50")
```

## Problem 8: ATM Withdrawal Simulation

**Question:** Simulate ATM rules: multiples of 100, sufficient balance, manager approval.

**Approach:** Check errors first, then nested conditions for amount > 10000.

Inputs:
- balance
- amount

Logic:
- If amount % 100 != 0 → reject (ATM works with ₹100 multiples)
- If amount > balance → reject
- Else:
  - if amount > 10000 → ask for manager approval
  - else → withdraw success

```python
balance = int(input("Enter your account balance: "))
amount = int(input("Enter amount to withdraw: "))

# Check invalid conditions first
if amount % 100 != 0:
    print("Transaction Failed: Amount must be in multiples of 100.")
elif amount > balance:
    print("Transaction Failed: Insufficient balance.")
else:
    # Valid amount and sufficient balance → now check for manager approval
    if amount > 10000:
        print("Manager approval required for this transaction.")
    else:
        print("Withdrawal successful.")
```

## Problem 9: Employee Performance Rating

**Question:** Assign performance rating based on sales, feedback, attendance.

**Approach:** Prioritize attendance rule, then apply nested conditions.

Inputs:
- sales (₹)
- feedback (0-5 scale)
- attendance (%)

Logic:
- If attendance < 70 → 'Poor'
- Else:
  - If sales > 100000 AND feedback > 4.5 → 'Excellent'
  - If sales > 70000 OR feedback > 4 → 'Good'
  - Else → 'Average'

```python
sales = float(input("Enter total sales amount: "))
feedback = float(input("Enter feedback rating (0–5): "))
attendance = float(input("Enter attendance percentage: "))

# Attendance rule gets highest priority
if attendance < 70:
    print("Performance Rating: Poor")
else:
    # Now check nested performance conditions
    if sales > 100000 and feedback > 4.5:
        print("Performance Rating: Excellent")
    elif sales > 70000 or feedback > 4:
        print("Performance Rating: Good")
    else:
        print("Performance Rating: Average")
```
